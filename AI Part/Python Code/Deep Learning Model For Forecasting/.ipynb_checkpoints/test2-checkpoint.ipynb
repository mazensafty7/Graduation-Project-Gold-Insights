```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

import tensorflow

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import TimeSeriesSplit
from tensorflow.keras.regularizers import l2
import random
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dropout, Dense
from keras.losses import Huber
from tensorflow.keras.layers import LayerNormalization
```

```python
df = pd.read_csv('Processed_Data.csv')
```

```python
df.head()
```

```python
df = df.rename(columns = {"Unnamed: 0" : "Date"})
```

```python
df["Date"] = pd.to_datetime(df["Date"] , format = "%Y-%m-%d")
```

```python
df.head()
```

```python
df = df.set_index("Date")
```

```python
df = df.reset_index(drop = True)
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap between Features')
plt.show()
```

```python
df = df.drop(columns=['US Dollar', 'Unrate'])
```

```python
train_size = int(len(df) * 0.8)
train_df = df.iloc[:train_size]
test_df = df.iloc[train_size:]
```

```python
scaler_f = MinMaxScaler()
scaler_f.fit(train_df.drop(columns=['Gold']))
```

```python
scaler_t = MinMaxScaler()
scaler_t.fit(train_df[['Gold']])
```

```python
train_features_scaled = scaler_f.transform(train_df.drop(columns=['Gold']))
train_gold_scaled = scaler_t.transform(train_df[['Gold']])
```

```python
test_features_scaled = scaler_f.transform(test_df.drop(columns=['Gold']))
test_gold_scaled = scaler_t.transform(test_df[['Gold']])
```

```python
train_scaled = pd.DataFrame(train_features_scaled, columns=train_df.columns.drop('Gold'), index=train_df.index)
train_scaled['Gold'] = train_gold_scaled.flatten()
```

```python
cols = ['Gold'] + [col for col in train_scaled.columns if col != 'Gold']
train_scaled = train_scaled[cols]
```

```python
train_scaled.head()
```

```python
test_scaled = pd.DataFrame(test_features_scaled, columns=test_df.columns.drop('Gold'), index=test_df.index)
test_scaled['Gold'] = test_gold_scaled.flatten()
```

```python
cols_test = ['Gold'] + [col for col in test_scaled.columns if col != 'Gold']
test_scaled = test_scaled[cols_test]
```

```python
test_scaled.head()
```

```python
def create_sequences(data, target_column='Gold', window_size=25):
    X = []
    y = []
    
    for i in range(window_size, len(data)):
        X.append(data.iloc[i-window_size:i].values)  # كل الأعمدة كـ features
        y.append(data.iloc[i][target_column])        # الهدف هو الذهب فقط

    return np.array(X), np.array(y)
```

```python
X_train, y_train = create_sequences(train_scaled)
X_test, y_test = create_sequences(test_scaled)
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
def define_model():
    input1 = Input(shape=(25, 6))
    
    x = LSTM(128, return_sequences=True, kernel_regularizer=l2(0.001))(input1)
    x = BatchNormalization()(x)
    x = Dropout(0.4)(x)

    x = LSTM(32, return_sequences=True, kernel_regularizer=l2(0.001))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)

    x = LSTM(32, kernel_regularizer=l2(0.001))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)

    x = Dense(32, activation='linear')(x)
    dnn_output = Dense(1)(x)

    model = Model(inputs=input1, outputs=[dnn_output])
    model.compile(loss='mean_squared_error', optimizer=Adam(learning_rate=0.001))
    model.summary()

    return model
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=4, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-6)
```

```python
model = define_model()
history = model.fit(
    X_train, y_train,
    epochs=150,
    batch_size=32,
    validation_split=0.1,
    verbose=1,
    callbacks=[early_stop, reduce_lr],
)
```

```python
y_pred = model.predict(X_test)
```

```python
y_test_true = scaler_t.inverse_transform(y_test.reshape(-1, 1)).flatten()
y_test_pred = scaler_t.inverse_transform(y_pred).flatten()
```

```python
mse = mean_squared_error(y_test_true, y_test_pred)
rmse = mean_squared_error(y_test_true, y_test_pred, squared=False)
mae = mean_absolute_error(y_test_true, y_test_pred)
mape = mean_absolute_percentage_error(y_test_true, y_test_pred)
accuracy = 1 - mape
r2_test = r2_score(y_test_true, y_test_pred)

print(f"Test Loss (MSE): {mse:.4f}")
print(f"Test MAE: {mae:.4f}")
print(f"Test MAPE: {mape:.4f}")
print(f"Test Accuracy: {accuracy * 100:.2f}%")
print(f"Test RMSE: {rmse:.2f}")
print(f"Test R² Score: {r2_test * 100:.2f}%")
```

```python
# Plot results
plt.figure(figsize=(14, 7))
plt.plot(y_test_true, label='Actual Gold Price', color='blue', alpha=0.8)
plt.plot(y_test_pred, label='Predicted Gold Price', color='orange', alpha=0.8)
plt.title('Gold Price Prediction with TimeSeriesSplit')
plt.xlabel('Time Step')
plt.ylabel('Gold Price')
plt.legend()
plt.grid(True)
plt.show()
```
